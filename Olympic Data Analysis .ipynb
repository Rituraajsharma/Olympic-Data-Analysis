# Handling missing values

```python
df.fillna(np.nan, inplace=True)  # Replace empty strings with NaN
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')  # Convert Age to numeric
df['Height'] = pd.to_numeric(df['Height'], errors='coerce')  # Convert Height to numeric
df['Weight'] = pd.to_numeric(df['Weight'], errors='coerce')  # Convert Weight to numeric
```

# Drop rows with missing values in important columns

```python
df.dropna(subset=['Age', 'Height', 'Weight', 'Medal'], inplace=True)
```

# Analyzing and visualizing with question

# Q.1  How does age affect medal wins for different genders?

```python
medal_counts = df.groupby('Medal')['Medal'].count()
medal_counts
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x=medal_counts.index, y=medal_counts.values)
plt.title('Distribution of Medals')
plt.xlabel('Medal Type')
plt.ylabel('Number of Medals')
plt.show()
```

# Q.2 Which countries have won the most medals overall?

```python
medal_df = df[df['Medal'].notna()]
```

```python
medal_counts = medal_df.groupby(['NOC', 'Medal'])['Medal'].count().reset_index(name='Count')
```

```python
top_countries = medal_counts.groupby('NOC')['Count'].sum().sort_values(ascending=False).head(10).index
top_countries
```

```python
top_countries_df = medal_counts[medal_counts['NOC'].isin(top_countries)]
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='NOC', y='Count', hue='Medal', data=top_countries_df)
plt.title('Medal Distribution for Top 10 Countries')
plt.xlabel('Country (NOC)')
plt.ylabel('Number of Medals')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.legend(title='Medal')
plt.show()
```

# Q.3 How has the average age of athletes changed over the years?

```python
average_age_by_year = df.groupby('Year')['Age'].mean().reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Age', data=average_age_by_year)
plt.title('Average Age of Athletes Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Age')
plt.grid(True)  # Adding a grid for better readability
plt.show()
```

#Q.4 What are the most popular sports in the Olympics based on the number of participants?

```python
sport_counts = df.groupby('Sport')['Name'].count().sort_values(ascending=False)
```

```python
top_10_sports = sport_counts.head(10)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_sports.index, y=top_10_sports.values)
plt.title('Top 10 Most Popular Olympic Sports by Number of Participants')
plt.xlabel('Sport')
plt.ylabel('Number of Participants')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

# Q.5 Which sports have the highest proportion of female athletes?

```python
sport_gender_counts = df.groupby(['Sport', 'Sex'])['Name'].count().reset_index(name='Count')
sport_totals = sport_gender_counts.groupby('Sport')['Count'].sum().reset_index(name='Total')
sport_gender_proportions = pd.merge(sport_gender_counts, sport_totals, on='Sport')
sport_gender_proportions['Proportion'] = sport_gender_proportions['Count'] / sport_gender_proportions['Total']
```

```python
female_proportions = sport_gender_proportions[sport_gender_proportions['Sex'] == 'F'].sort_values(by='Proportion', ascending=False)
```

```python
top_10_female_sports = female_proportions.head(10)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Sport', y='Proportion', data=top_10_female_sports)
plt.title('Top 10 Sports with the Highest Proportion of Female Athletes')
plt.xlabel('Sport')
plt.ylabel('Proportion of Female Athletes')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# Q.6 How does the height and weight of athletes vary across different sports?

```python
sports = ['Basketball', 'Gymnastics', 'Swimming', 'Weightlifting']
filtered_df = df[df['Sport'].isin(sports)]
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Sport', y='Height', data=filtered_df)
plt.title('Height Variation Across Different Sports')
plt.xlabel('Sport')
plt.ylabel('Height (cm)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Sport', y='Weight', data=filtered_df)
plt.title('Weight Variation Across Different Sports')
plt.xlabel('Sport')
plt.ylabel('Weight (kg)')
plt.show()
```
